Desarrolla un programa en Python que funcione como una calculadora geométrica interactiva. El sistema debe mostrar un menú cíclico que permita al usuario seleccionar entre calcular el área de un círculo, su perímetro o el volumen de una esfera. Cada opción solicitará al usuario el valor del radio en el momento de ejecutarse. El programa deberá mantenerse en funcionamiento hasta que el usuario decida salir. Este ejercicio tiene como objetivo aplicar la programación procedural, el uso de estructuras de control repetitivas y condicionales, así como la organización del código mediante funciones reutilizables.

1. Área del círculo =  A=π⋅r2A=π⋅r 2
2. Perímetro del círculo =  P=2⋅π⋅rP=2⋅π⋅r
3. Volumen de la esfera =  V=43⋅π⋅r3V= 34⋅π⋅r 3

```python
# ---------------------------------------------------
# PROGRAMA DE CALCULO GEOTRICO
# Elaborado por Jonathan baray
# Formulas utilizadas:
# 1. Área del círculo =  A=π⋅r2A=π⋅r 2
2. Perímetro del círculo =  P=2⋅π⋅rP=2⋅π⋅r
3. Volumen de la esfera =  V=43⋅π⋅r3V= 34⋅π⋅r 3
# --------------------------------------------------

import math

def areacirculo():
    """Calcula el área de un círculo usando la fórmula A = π * r²"""
    print("\n📐 CÁLCULO DEL ÁREA DEL CÍRCULO")
    radio = float(input("👉 Ingresa el radio del círculo: "))
    area = math.pi * (radio ** 2)
    print(f"✅ El área es: {area:.2f} unidades cuadradas\n")

def perimetrocirculo():
    """Calcula el perímetro de un círculo usando la fórmula P = 2 * π * r"""
    print("\n📏 CÁLCULO DEL PERÍMETRO DEL CÍRCULO")
    radio = float(input("👉 Ingresa el radio del círculo: "))
    perimetro = 2 * math.pi * radio
    print(f"✅ El perímetro es: {perimetro:.2f} unidades lineales\n")

def volumenesfera():
    """Calcula el volumen de una esfera usando la fórmula V = (4/3) * π * r³"""
    print("\n🌍 CÁLCULO DEL VOLUMEN DE LA ESFERA")
    radio = float(input("👉 Ingresa el radio de la esfera: "))
    volumen = (4/3) * math.pi * (radio ** 3)
    print(f"✅ El volumen es: {volumen:.2f} unidades cúbicas\n")

def menu():
    """Función principal que muestra el menú y controla el flujo del programa"""
    while True:
        print("\n✨🔢 MENÚ DE CÁLCULOS GEOMÉTRICOS 🔢✨")
        print("1️⃣  Calcular área del círculo")
        print("2️⃣  Calcular perímetro del círculo")
        print("3️⃣  Calcular volumen de la esfera")
        print("4️⃣  Salir")

        opcion = int(input("👉 Selecciona una opción (1-4): "))

        if opcion == 1:
            areacirculo()
        elif opcion == 2:
            perimetrocirculo()
        elif opcion == 3:
            volumenesfera()
        elif opcion == 4:
            print("\n👋 ¡Gracias por usar la calculadora! Hasta la próxima. 🚀\n")
            break
        else:
            print("⚠️ Opción no reconocida. Por favor, elige un número del 1 al 4.")

# 🔹 Punto de entrada principal
if __name__ == "__main__":
    menu()
```
